# FFT with $C^1$ continuous extension  

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import square
```

## Classical FS expansion

$$ f(x) = c + \sum_n^\infty a_n cos(n\omega x) + \sum_n^\infty b_n sin(n\omega x), \quad x \in [x_l, x_u]$$
where $f(x) = f(x + T)$ is assumed periodic with period $T = x_u - x_l$.

```python
# Assume x (independent variable) and y are the data.
# Arbitrary numerical values for question purposes:
xl = 0
xu = 1
xuu = 0.5*(3*xu- xl)
terms = 5 # number of terms for the Fourier series

x0 = np.linspace(xl, xu, 400, endpoint=True) 
func = lambda x: x
y = func(x0)
# Period
tau = xu - xl # assume that signal length = 1 period (tau)

def fourier_approx(y, terms):
    fourier_coeffs = np.fft.rfft(y) 
    np.put(fourier_coeffs, range(terms+1, len(fourier_coeffs)), 0.0) # zero-ing coefficients above "terms"
    return np.fft.irfft(fourier_coeffs)

y_inverse_fourier = fourier_approx(y, terms)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax.plot(x0, y, label = 'artificial data')
ax.plot(x0, y_inverse_fourier, label = ("'FFT' series fit (%d terms)" % terms))
ax.grid(True, color='dimgray', linestyle='--', linewidth=0.5)
ax.set_axisbelow(True)
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_title('Gibbs effect')
ax.legend()
```

# $C^1$ extension of $f(x)$

We form $F$ over the wanted, original support $[ x_l, x_u]$

$$F(x) = \begin{cases}
          f(x) \quad &\text{if} \, x \in [x_l, x_u] \\
          p(x) \quad &\text{if} \, x \in [x_u, \tilde{x}_u]
     \end{cases}
     $$

```python
cubic_base = lambda x: x**np.arange(4)
cubic_base_derivative = lambda x: np.arange(4)*x**np.arange(-1, 3)

def periodic_extension(func, func_derivative, xl, xu, xuu, extension_only=False):
    xl = float(xl)
    xu = float(xu)
    xuu = float(xuu)
    system_matrix = np.vstack([cubic_base(xuu), 
                               cubic_base_derivative(xuu),
                               cubic_base(xu),
                               cubic_base_derivative(xu)])
    
    constraints = np.array([func(xl), func_derivative(xl), func(xu), func_derivative(xu)])
    cubic_coeffs = np.linalg.solve(system_matrix, constraints)
    cubic_value = lambda x: cubic_coeffs.dot(cubic_base(x))
    cubic_extension = np.vectorize(cubic_value)
    if extension_only:
        return cubic_extension
    else:
        return (lambda x: np.piecewise(x, [x <= xu, x>xu], [func, cubic_extension]))
```

```python
Func = periodic_extension(lambda x: x, lambda x: 1, xl, xu, xuu)
x = np.linspace(xl, xuu, 400)
# Func = np.piecewise(x, [x <= xu, x>xu], [func, cubic_extension])
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x, Func(x), label = 'Extended f(x)')
ax.plot(x, fourier_approx(Func(x), terms), label = ("FFT extended series fit (%d terms)" % terms))
ax.grid(True, color='dimgray', linestyle='--', linewidth=0.5)
ax.set_axisbelow(True)
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.legend()
```
